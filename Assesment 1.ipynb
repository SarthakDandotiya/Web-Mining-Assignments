### Question 1
Write a program to extract the source content (including html tags) from a website.
Filter the html tags from the content in order to extract the plain text from them and
display the number of words and term frequency of each word present in them after
applying stop word removal. Also, apply stemming to the same document and display
the number of words along with their corresponding stemmed words present in them.


```python
import urllib.request
from bs4 import BeautifulSoup
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.probability import FreqDist
import pandas as pd

url = "https://sarthakdandotiya.github.io/"
response = urllib.request.urlopen(url)
html = response.read()
soup = BeautifulSoup(html)
ps = PorterStemmer()

# Source Content
# print(soup)

# Getting rid of the info within script tags
for script in soup(["script", "style"]):
    script.extract() 

# Extracting Text:
text = soup.get_text()

# Tokenization
tokenized = word_tokenize(text)
# Convert everything to lower case
tokenized = [x.lower() for x in tokenized]
print("INITIALLY -")
print("Total Words:", len(tokenized), end="\n\n")
    
# Frequencies of each word
def display_freq(wordlist):
    freq_vec = FreqDist(wordlist)
    print("Total Words:",len(freq_vec))

def freq_df(wordlist):
    freq_vec = FreqDist(wordlist)
    word = []
    stemmed = []
    count = []
    for k, v in freq_vec.items():
        word.append(k)
        stemmed.append(ps.stem(k))
        count.append(v)
#         print(k, "-", ps.stem(k), ": ", v)
    data = { "Words": word, "Stemmed": stemmed, "Count":count }
    df = pd.DataFrame(data, columns=["Words", "Stemmed", "Count"])
    return df

# Stopword Removal
my_stopwords = set(stopwords.words('english'))
filtered = [w for w in tokenized if not w in my_stopwords]

print("AFTER STOPWORD REMOVAL -")
display_freq(filtered)

# Stemming
stemmed = []
for token in filtered:
    stemmed_word = ps.stem(token)
    stemmed.append(stemmed_word)

print("\n\nAFTER STEMMING -")
display_freq(stemmed)

print("\n\nWrods with their corresponding Stemmed word & Count -\n")
freq_df(filtered)
```

### Question 3
Write a program to extract the contents (excluding the html tags) from two
educational websites and save the content in two separate .doc file. Remove
stopwords from the content and represent the documents using Boolean, Bag-of-words and Complete representation.

```python
import urllib.request
from bs4 import BeautifulSoup
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
from docx import Document
import numpy as np
import collections 
import pandas as pd
from IPython.display import display

url1 = "https://www.geeksforgeeks.org/types-of-operating-systems/"
response1 = urllib.request.urlopen(url1)
html1 = response1.read()
soup1 = BeautifulSoup(html1)

url2 = "https://www.tutorialspoint.com/operating_system/os_types.htm"
response2 = urllib.request.urlopen(url2)
html2 = response2.read()
soup2 = BeautifulSoup(html2)

# Getting rid of the info within script tags
for script in soup1(["script", "style"]):
    script.extract() 

for script in soup2(["script", "style"]):
    script.extract() 

# Extracting Text:
text1 = soup1.get_text()
text2 = soup2.get_text()

# Saving into docx files
document1 = Document()
document2 = Document()
paragraph1 = document1.add_paragraph(text1)
paragraph2 = document2.add_paragraph(text2)
document1.save('A1-first.docx')
document2.save('A1-second.docx')

# Tokenization
tokenized1 = word_tokenize(text1)
tokenized2 = word_tokenize(text2)
# Convert everything to lower case
tokenized1 = [x.lower() for x in tokenized1]
tokenized2 = [x.lower() for x in tokenized2]

# Stopword Removal
my_stopwords = set(stopwords.words('english'))
# Adding Custom Stop-word
my_stopwords.update([
    ".", ",", "'", ":", ";", "\"", "...",
    "(", ")", "{", "}", "[", "]", "©", "@",
    "'m", "--", "__", "?", "|", "#", "!",
    "-", "$", "'s", "&", "·", "–", "’", "‘",
    "“", "”", "%", "''", "+", "/","-",])
filtered1 = [w for w in tokenized1 if not w in my_stopwords]
filtered2 = [w for w in tokenized2 if not w in my_stopwords]

# print(filtered1)
# print(collections.Counter(filtered1))

# Boolean Representation...
def boolean(original, modified):
#     original -> tokenized input text
#     modified -> tokens after removing stopwords/stemming
    bag = np.zeros(len(original), dtype = int)
    for i,original_word in enumerate(original):
        if original_word in modified:
            bag[i]=1
    return np.array(bag)

# Bag of Word Reresentation...
def bow(original, modified):
    bag = np.zeros(len(original), dtype = int)
    for i,sw in enumerate(original):
        for j,word in enumerate(modified):
            if word == sw: 
                bag[i] += 1
    return np.array(bag)

# Removing duplicates
def Remove(duplicate): 
    final_list = [] 
    for num in duplicate: 
        if num not in final_list: 
            final_list.append(num) 
    return final_list

        
def convert2df(zipped, label):
    words = []
    count = []
    for i,x in enumerate(zipped):
        words.append(x[0])
        count.append(x[1])
    data = { "Words": words, label :count }
    df = pd.DataFrame(data, columns=["Words", label])
    return df

def convert2df_complete(zipped):
    words = []
    start = []
    end = []
    for i,x in enumerate(zipped):
        words.append(x[0])
        start.append(x[1])
        end.append(x[2])
    data = { "Words": words, "Starts At" : start, "Ends At": end }
    df = pd.DataFrame(data, columns=["Words", "Starts At", "Ends At"])
    return df

print("\n------------------------- BOOLEAN for Document 1 -------------------------\n")
temp_boo1 = boolean(tokenized1, filtered1)
boo1 = Remove(list(zip(tokenized1, temp_boo1)))
display(convert2df(boo1, "Boolean"))

print("\n------------------------- BOOLEAN for Document 2 -------------------------\n")
temp_boo2 = boolean(tokenized2, filtered2)
boo2 = Remove(list(zip(tokenized2, temp_boo2)))
display(convert2df(boo2, "Boolean"))

print("\n------------------------- BOW for Document 1 -------------------------\n")
temp_bow1 = bow(tokenized1, filtered1)
bow1 = Remove(list(zip(tokenized1, temp_bow1)))
display(convert2df(bow1, "Bag Of Words"))

print("\n------------------------- BOW for Document 2 -------------------------\n")
temp_bow2 = bow(tokenized2, filtered2)
bow2 = Remove(list(zip(tokenized2, temp_bow2)))
display(convert2df(bow2, "Bag Of Words"))

def complete(txt):
    tokens = word_tokenize(txt)
    offset = 0
    for token in tokens:
        offset = txt.find(token, offset)
        yield token, offset, offset+len(token)
        offset += len(token)

print("\n------------------------- COMPLETE REPRESENTATION for Document 1 -------------------------\n")
C1 = []
for token in complete(text1):
#     print(token)
    C1.append(token)
    token[0]==text1[token[1]:token[2]]
display(convert2df_complete(C1))
    
print("\n------------------------- COMPLETE REPRESENTATION for Document 2 -------------------------\n")
C2 =[]
for token in complete(text2):
#     print(token)
    C2.append(token)
    token[0] == text2[token[1]:token[2]]
display(convert2df_complete(C2))
```

### Question 4
 Process a search a query and compare the contents of the both pages collected in
Q.No. 3 with the processed query, display the similarity result based on highest
matching count (bag-of-words)

```python
search = word_tokenize("operating system types")

def similarity(query,filtered):
    vals = []
    c = collections.Counter(filtered)
    for w in query:
        if w in c:
            vals.append(c[w])
    return sum(vals)

si1 = similarity(search,filtered1)
si2 = similarity(search,filtered2)

max = si1 if si1 > si2 else si2

print("Similarity:")
print("Doc1 or Website1 :", si1)
print("Doc2 or Website2 :", si2)

if (max == si1):
    print("\nDoc 1 has a better match")
elif (max == si2):
    print("\nDoc 2 has a better match")
```
